```python
#Read csv as dataframe 
import csv
import pandas as pd
csv_name= 'stroopdata.csv'
data = pd.read_csv(csv_name)
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
congruent=data['Congruent']
incongruent=data['Incongruent']
```

```python
from scipy.stats import norm
(mu, sigma) = norm.fit(congruent)
(m,sig)=norm.fit(incongruent)
```

```python
plt.rcParams.update({'font.size': 22})
fig= plt.figure(figsize=(20,12))
n, bins, patches = plt.hist(congruent,10, normed=1, facecolor='green', alpha=0.75,label='Congruent')
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=4)
#plt.hist(congruent,bins=4,alpha=0.7, label='Congruent')
#plt.hist(incongruent, bins=4,alpha=0.7, label='Incongruent')
plt.legend(loc='upper right')
plt.xlabel('Response Time')
plt.ylabel('Frequency')
plt.show()
fig.savefig('Congruent.png')
```

```python
plt.rcParams.update({'font.size': 22})
fig= plt.figure(figsize=(20,12))
n, bins, patches = plt.hist(incongruent,10, normed=1, facecolor='blue', alpha=0.75,label='Incongruent')
y = mlab.normpdf( bins, m, sig)
l = plt.plot(bins, y, 'k--', linewidth=4)
#plt.hist(congruent,bins=4,alpha=0.7, label='Congruent')
#plt.hist(incongruent, bins=4,alpha=0.7, label='Incongruent')
plt.legend(loc='upper right')
plt.xlabel('Response Time')
plt.ylabel('Frequency')
plt.show()
fig.savefig('Incongruent.png')
```
